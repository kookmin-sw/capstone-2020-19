## 낙상 데이터 343, 걷기 데이터 289 
-> 280개 기준으로 가속도 차이 확인.

1. 일단 전체 데이터들의 평균 파일 길이를 확인한다. 

*2. 평균 파일 길이 이상인 파일들만 남긴다. (**너무 짧은 데이터 제거**) -> 전체 데이터 확보된 이후에 하기로*
3. 그 파일들에 대해서 x, y, z 들의 센서값을 제곱해서 더한후, 길이로 나눠준다. (x, y, z각각)
    - 이 값들을 리스트에 담는다. 
4. 리스트 평균을 낸다. (낙상/ 걷기 각각.)(**평균 센서값으로 분류 가능한지 확인**)

```python
import os 

w_data_len = []
f_data_len = []
path = ['/home/ghyeon/ML/w_data', '/home/ghyeon/ML/f_data']
# count = 0 

for i in range(2):
    for file in os.listdir(path[i]):
        filename = os.fsdecode(file)
        if filename.endswith(".csv") and i == 0: 
            data = pd.read_csv('/home/ghyeon/ML/w_data/' + filename, names=colname)
            d_len = len(data['x'][1:])
            w_data_len.append(d_len)
        if filename.endswith(".csv") and i == 1:
            data = pd.read_csv('/home/ghyeon/ML/f_data/' + filename, names=colname)
            d_len = len(data['x'][1:])
            f_data_len.append(d_len)
        else:
            continue

w_avg = sum(w_data_len) // len(w_data_len)
f_avg = sum(f_data_len) // len(f_data_len)
```

```python
print("걷기 데이터 평균 길이 = ", w_avg)
print(len([i for i in w_data_len if i < 40]), " (걷기) 만큼의 데이터가 40보다 짧습니다. ")
print(len([i for i in w_data_len if i < w_avg]), " (걷기) 만큼의 데이터가 평균보다 짧습니다. \n")
print("낙상 데이터 평균 길이 = ", f_avg)
print(len([j for j in f_data_len if j < 30]), " (낙상) 만큼의 데이터가 30보다 짧습니다. ")
print(len([j for j in f_data_len if j < f_avg]), " (낙상) 만큼의 데이터가 평균보다 짧습니다. ")
```

**걷기 데이터 평균 길이 : 243, 낙상 데이터 평균 길이 : 43**

적절한 데이터가 아닌지, 실수로 들어온 데이터인지를 적정 수준의 길이로 걸러내기로 함. 
- **걷기 데이터 제거 기준 길이 : 40, 낙상 데이터 제거 기준 길이 30**

```python
from math import pow

square_root_f = []
square_root_w = []
path = ['/home/ghyeon/ML/w_data', '/home/ghyeon/ML/f_data']

# 각 파일마다의 x, y, z값을 제곱후 전부 더하고 길이로 나눠주면 된다. 
# [x, y, z]를 가지게 됨. 

for i in range(2):
    for file in os.listdir(path[i]):
        filename = os.fsdecode(file)
        if filename.endswith(".csv") and i == 0: 
            # walk
            data = pd.read_csv('/home/ghyeon/ML/w_data/' + filename)
            list_x = data['x'][1:]
            list_y = data['y'][1:]
            list_z = data['z'][1:]
            # zero division error.
            try:
                sq_r_x = sum([pow(float(x), 2) for x in list_x]) // len(list_x)
                sq_r_y = sum([pow(float(x), 2) for x in list_y]) // len(list_y)
                sq_r_z = sum([pow(float(x), 2) for x in list_z]) // len(list_z)
                square_root_w.append([sq_r_x, sq_r_y, sq_r_z])
            except:
                continue
        if filename.endswith(".csv") and i == 1:
            # fall
            data = pd.read_csv('/home/ghyeon/ML/f_data/' + filename)
            list_x = data['x'][1:]
            list_y = data['y'][1:]
            list_z = data['z'][1:]
            # zero division error.
            try:
                sq_r_x = sum([pow(float(x), 2) for x in list_x]) // len(list_x)
                sq_r_y = sum([pow(float(x), 2) for x in list_y]) // len(list_y)
                sq_r_z = sum([pow(float(x), 2) for x in list_z]) // len(list_z)
                square_root_f.append([sq_r_x, sq_r_y, sq_r_z])
            except:
                continue
        else:
            continue
```

```python
print("낙상 데이터 평균, x, y, z ")
print(sum([x for x, y, z in square_root_f]) / len(square_root_f))
print(sum([y for x, y, z in square_root_f]) / len(square_root_f))
print(sum([z for x, y, z in square_root_f]) / len(square_root_f))
print("")
print("걷기 데이터 평균, x, y, z")
print(sum([x for x, y, z in square_root_w]) / len(square_root_w))
print(sum([y for x, y, z in square_root_w]) / len(square_root_w))
print(sum([z for x, y, z in square_root_w]) / len(square_root_w))
```
