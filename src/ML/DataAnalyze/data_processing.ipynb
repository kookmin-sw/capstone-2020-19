# Problem : OVERfitting. 

0.0001정도밖에 차이나지 않는 데이터셋을 전부 사용해서 학습하려고 하니 정확도가 100씩 나온다. 오버피팅이 심하게 됐음. 

augmentation 을 하는 대신에, 데이터셋을 늘리기 위해서 30의 길이 이상인 경우 데이터셋을 나눠서 새로운 csv파일로 저장해주는 과정을 하려고 한다.
