## 데이터 augmentation 
- x, y, z의 데이터들에서 -0.0001 / +0.0001 을 하거나 하지 않는 식으로 하나의 파일을 26개 더 늘린다.
- x : o + - / y : o + - / z : o + - -> 27개 경우의 수. o o o 로 원래의 파일과 똑같은것 말고 나머지파일은 새로 생기는것임.

- unbalanced data 고쳐주느라 f_data 에는 뻥튀기함수를 적용했음. 
- 7000/ 7000 개로 -> 5000 / 2000 -> 10000 / 4000개. 

train : w_data 5000 + f_data 5000

test  : w_data 2000 + f_data 2000

```python
import pandas as pd 
import os
from pandas import DataFrame
import random 

m_n = [i for i in range(30)]

test_row_id = []
test_series_id = []
test_measure_num = []
test_x = []
test_y = []
test_z = []
test_count = 0

train_row_id = []
train_series_id = []
train_measure_num = []
train_x = []
train_y = []
train_z = []
train_count = 0

f_file_name = os.listdir('/home/ghyeon/ML/f_data')
random.shuffle(f_file_name)
w_file_name = os.listdir('/home/ghyeon/ML/w_data')
random.shuffle(w_file_name)

for filename in f_file_name:
    if filename.endswith(".csv"):
        data = pd.read_csv('/home/ghyeon/ML/f_data/' + filename)
        x = list(data['x'])
        y = list(data['y'])
        z = list(data['z'])
        if len(x) < 30:
            continue
        elif len(x) >= 30 and train_count < 5000:
            train_x.extend(x[:30])
            train_y.extend(y[:30])
            train_z.extend(z[:30])
            train_series_id.extend([train_count] * 30)
            train_measure_num.extend(m_n)
            train_row_id.extend(['{}_{}'.format(train_count, i) for i in range(30)])
            train_count += 1
        elif len(x) >= 30 and test_count < 2000:
            test_x.extend(x[:30])
            test_y.extend(y[:30])
            test_z.extend(z[:30])
            test_series_id.extend([test_count] * 30)
            test_measure_num.extend(m_n)
            test_row_id.extend(['{}_{}'.format(test_count, i) for i in range(30)])
            test_count += 1

print(test_count)
print(train_count)
            
for filename in w_file_name:
    if filename.endswith(".csv"):
        data = pd.read_csv('/home/ghyeon/ML/w_data/' + filename)
        x = list(data['x'])
        y = list(data['y'])
        z = list(data['z'])
        if len(x) < 30:
            continue
        elif len(x) >= 30 and train_count < 10000:
            train_x.extend(x[:30])
            train_y.extend(y[:30])
            train_z.extend(z[:30])
            train_series_id.extend([train_count] * 30)
            train_measure_num.extend(m_n)
            train_row_id.extend(['{}_{}'.format(train_count, i) for i in range(30)])
            train_count += 1
        elif len(x) >= 30 and test_count < 4000:
            test_x.extend(x[:30])
            test_y.extend(y[:30])
            test_z.extend(z[:30])
            test_series_id.extend([test_count] * 30)
            test_measure_num.extend(m_n)
            test_row_id.extend(['{}_{}'.format(test_count, i) for i in range(30)])
            test_count += 1
        
print(test_count)
print(train_count)

train_data_dict = {'row_id' : train_row_id, 'series_id' : train_series_id, 'measurement_num' : train_measure_num,
             'x' : train_x, 'y' : train_y, 'z' : train_z}
train_data_df = DataFrame(train_data_dict)

test_data_dict = {'row_id' : test_row_id, 'series_id' : test_series_id, 'measurement_num' : test_measure_num,
             'x' : test_x, 'y' : test_y, 'z' : test_z}
test_data_df = DataFrame(test_data_dict)
```
